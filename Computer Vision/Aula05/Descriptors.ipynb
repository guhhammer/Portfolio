In the scikit image manual, we can find several descriptors

https://scikit-image.org/docs/dev/api/skimage.feature.html?highlight=template

Below we illustrate the daisy and the pca. Take a look:

```python
%matplotlib inline
import matplotlib.pyplot as plt
from skimage.feature import daisy
from skimage import data
from skimage.transform import rescale
import math

img = data.camera()

desc = daisy(img, step=180, radius=58, rings=2, histograms=6,
                         orientations=8, visualize=False)

daise_feat = desc.flatten()
print(desc.shape,daise_feat.shape)
print(daise_feat[:20])
plt.imshow(img,cmap='gray')
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
pca_mat = pca.fit_transform(img)

pca_feat = pca_mat.flatten()
print(pca_mat.shape,pca_feat.shape)
print(pca_feat[:20])
```
