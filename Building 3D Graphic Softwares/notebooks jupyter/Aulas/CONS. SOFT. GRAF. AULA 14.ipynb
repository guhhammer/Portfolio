```python
def rgb2gray(rgb):
  fil = [0.299, 0.587, 0.144]
  return np.dot(rgb, fil)
```

```python
import scipy.signal as scs

filtered = scs.medfilt(img1)
py.imshow(filtered)
```

```python
#Seleção de pontos segundo a tonalidade.
imgmod = py.imread('giraffemod.png')
filtered = scs.medfilt(imgmod)
reddish = filtered[:, :, 0] > 200/255 # reddish é uma matriz booleana
fx = filtered.copy() # copiar um array (sem alterar o original)
fx[reddish] = [1, 1, 0] # os pts True de fx são alterados
py.imshow(fx)
```

```python
from scipy import ndimage
# ndimage: módulo para rotação de imagens
```

```python
# ângulo de rotação: 10 graus
# mode: modo como os pixels situados fora do array original são tratados
# cval: valor atribuído aos pixels situados fora do array original
im_rot = ndimage.rotate(img,180,mode='constant',cval=0.8)
img2 = rgb2gray(im_rot)
py.imshow(img2,cmap='inferno')
# método de interpolação: nearest
#py.imshow(img2,interpolation='nearest')
```
