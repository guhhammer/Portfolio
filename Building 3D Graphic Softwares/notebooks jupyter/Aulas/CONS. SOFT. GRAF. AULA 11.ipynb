\end{equation*}

\begin{equation*} 
equação \ matricial: \ \ \ \ \ \ \

\end{vmatrix} \ = \ \begin{vmatrix}
\mathbf{x_0} \\ \mathbf{y_0} \\ \mathbf{z_0}
\end{vmatrix} \ + \ \gamma \ \begin{vmatrix}
\mathbf{s_x} \\ \mathbf{s_y} \\ \mathbf{s_z}

\mathbf{t_x} \\ \mathbf{t_y} \\ \mathbf{t_z}

\end{equation*}

\begin{equation*}
Obtenção \ das \ equações \ paramétricas: \ \ \ \ \ \ \

\mathbf{\gamma} \times \mathbf{s_x} \\ \mathbf{\gamma} \times \mathbf{s_y} \\ \mathbf{\gamma} \times \mathbf{s_z}

```python
def minus(v1, v2):
    return ((v1[0] - v2[0]), (v1[1] - v2[1]), (v1[2] - v2[2]))

def kons(k, v1):
    return ((k*(v1[0])),(k*(v1[1])),(k*(v1[2])))

def soma(v1,v2):
    return ((v1[0] + v2[0]), (v1[1] + v2[1]), (v1[2] + v2[2]))
    
def randdot(gamma, phi, v1, v2, v3):
    m1 = minus(v2,v1)
    m2 = minus(v3,v2)
    t1 = kons(gamma, m1)    #k1 -> gamma
    t2 = kons(phi, m2)    #k2 -> phi
    s1 = soma(v1, t1)
    return soma(s1, t2)

gamma = minus(P2,P1)
phi = minus(P3,P2)
print("P = {} + {}*{} + {}*{}".format(P1,'y', gamma, 'φ', phi),"\n")
print("X = {} + {}*{} + {}*{}".format(P1[0],'y', gamma[0], 'φ', phi[0]))
print("Y = {} + {}*{} + {}*{}".format(P1[1],'y', gamma[1], 'φ', phi[1]))
print("Z = {} + {}*{} + {}*{}".format(P1[2],'y', gamma[2], 'φ', phi[2]))


##print(randdot(g,p,P1,P2,P3))
```
