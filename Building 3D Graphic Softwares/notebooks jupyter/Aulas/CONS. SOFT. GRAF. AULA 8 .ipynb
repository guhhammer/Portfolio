### Equações do Plano
Equação do plano que passa pelo ponto $p0=(x0,y0,z0)$ e é paralelo a dois vetores independentes $s=(sx,sy,sz)$ e $t=(tx,ty,tz)$:<br>
$p=p0+us+wt$  <- equação do plano $p$ no $p_0$, $p_0$ é fixo e está no plano $P$<br>

Em que s≠kt (eles são vetores independentes) e u e w são variáveis escalares independentes.<br>

P0, s, t são vetores ou pontos fixos, são valores dados.<br>

O vetor p define os pontos do plano em que u e w variam independentemente.<br>

Quando u e w são variados, estamos definindo pela regra do polígono o vetor p do plano $P$.<br> 

$x = x_0 + u*s_x + w*t_x$<br>
$y = y_0 + u*s_y + w*t_y$<br>
$z = z_0 + u*s_z + w*t_z$<br>

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(1000, 0, 50)
xline = (zline)**3
yline = np.cos(zline) + np.sin(xline)
ax.plot3D(xline, yline, zline, 'red')

# Data for three-dimensional scattered points
zdata = 40 * np.random.random(100)
xdata = np.sin(zdata) + 10 * np.random.randn(100)
ydata = np.cos(zdata) + 0.0001 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Blues');
ax.view_init(10,-40)
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2+10))

x = np.linspace(-8, 8, 30)
y = np.linspace(-8, 8, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 62, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
def g(x, y):
    return np.sin(np.sqrt((x ** 4) + (y ** 4) -5))

x = np.linspace(-8, 8, 5)
y = np.linspace(-8, 8, 5)

X, Y = np.meshgrid(y, x)
Z = g(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 63, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(60,35)
```

2. três equações paramétricas para x, y, z.  

```python
from matplotlib.tri import Triangulation
# radius in x-y plane

theta = np.linspace(0, 4* np.pi, 90)
w = np.linspace(-0.1, 0.7, 50)
w, theta = np.meshgrid(w, theta)
phi = 0.5 * theta
r = 1 + w * np.cos(phi)


x = np.ravel(r * np.cos(theta))
y = np.ravel(r * np.sin(theta))
z = np.ravel(w * np.sin(phi))
tri = Triangulation(np.ravel(w), np.ravel(theta))

ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles,
                cmap='viridis', linewidths=0.8);

ax.set_xlim(-1, 1); ax.set_ylim(-1, 1); ax.set_zlim(-1, 1);
ax.view_init(30,35)
```
