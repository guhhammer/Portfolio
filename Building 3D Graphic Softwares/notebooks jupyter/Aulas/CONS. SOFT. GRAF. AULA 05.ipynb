. Equação vetorial de uma reta que passa pelo ponto p_0 e é paralela a um vetor t dado: <br><br>

<h2><center>
$P(u) = P_0$ + u . **t**                   
</center></h2>
<br>
Em que:<br>
   .  p(u) = (x,y,z) é um vetor que depende de $p_0$, u e t;<br>
   . $P_0$ = $(x_0,y_0,z_0)$ é um vetor de constante; <br>
   . u é uma **váriavel** escalar; <br>
   . t = $(t_x,t_y,t_z)$ é um vetor constante;< br>

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def somaV(v1,v2):
    somaV = []
    if len(v1) != len(v2):
        return "vectors must have the same length"
    return tuple([v1[i]+v2[i] for i in range(0,len(v1))])

def escalarV(k, v):
    return tuple([ k * v[i] for i in range(len(v))])


def mplot(v1,v2):
    plt.plot((v1[0], v2[0]),( v1[1], v2[1]))
```

```python
origem = (0,0)

p0 = (4,3)
t = (1,-3)

[-1.0,-0.5,0,0.5,1.0,1.5,2.0]
u = np.linspace(-1.0, 2.0, 6)
x0 = p0[0]
y0 = p0[1]

tx = t[0]
ty = t[1]

x = x0 + u*tx
y = y0 + u*ty

mplot(origem, p0)
mplot(origem, t)
plt.plot(x,y)            

plt.show()



```

```python
mplot(origem, p0)
mplot(origem, t)
mplot(origem, somaV(p0,t))
mplot(p0, somaV(p0,t))
```
